```python
def index(n_nodes, matrix):
    
    lines = []
    weights = np.random.randint(1, 1000, size=(1, 2))
    weights_sum = np.reshape(weights.sum(axis=1), (weights.shape[0],1))
    new_weights = weights/weights_sum
    A = np.dot(return_matrix,new_weights.T)
    index_ = np.argsort(A[:,0])
    
    for i in range(1, n_nodes+1):
        lines.append(index_[int(123*(i/n_nodes))-1])
        
    #print(lines)
    return matrix[lines]
```

```python
def bootstrap(n_nodes, matrix):
    
    return1 = 1
    return2 = 1
    index = np.random.randint(1, matrix.shape[0], size=12)
    
    for i in index:
        return1 = return1*(1+matrix[i][0])
        return2 = return2*(1+matrix[i][1])
    
    return return1, return2
```

```python
def tree(stages, n_nodes, matrix):
    
   
    tree_1 = {}
    tree_2 = {}
    tree_1[str(0)] = [100]
    tree_2[str(0)] = [100]
    for i in range(1, stages+1):
        aux1 = []
        aux2 = []
        #nodes = index(n_nodes, matrix)
        for j in range(n_nodes**(i-1)):
            for k in range(n_nodes):
                return1, return2 = bootstrap(n_nodes, matrix)
                aux1.append(tree_1[str(i-1)][j]*return1)
                aux2.append(tree_2[str(i-1)][j]*return2)
            tree_1[str(i)] = aux1
            tree_2[str(i)] = aux2
    
                
            
    return tree_1, tree_2     
```

```python
def liabilities(stages):
    
    lt = np.random.randint(300, 700, size=stages)
    #lt = np.zeros(stages)
    Lt = []
    
    
    for i in range(1, len(lt)+1):
        aux = 0
        for j in range(i, len(lt)+1):
            aux = aux + lt[j-1]/((1.05)**(j - i))
        
        Lt.append(aux)
    
    Lt = np.array(Lt)
    
    return lt, Lt
```

```python
def contributions(stages):
    
    ft = np.random.randint(300, 301, size=stages)
    #ft = np.zeros(stages)
    Ft = []
    
    
    for i in range(1, len(ft)+1):
        aux = 0
        for j in range(i, len(ft)+1):
            aux = aux + ft[j-1]/((1.05)**(j - i))
        
        Ft.append(aux)
    
    Ft = np.array(Ft)
    
    return ft, Ft
```

```python
def instance_creator(N, stages, n_nodes, M, Q, K, pi, matrix, n_instances):    
    
    for inst in range(1, n_instances+1):
        
        file = open("instancia" + str(n_nodes) + "-" + str(stages) + "-" + str(inst)+ ".txt", "wb")
        #file1 = open("instancia2.txt", "wb")


        file.write((str(inst)+"\n").encode('ascii'))
        file.write((str(N)+"\n").encode('ascii'))
        file.write((str(stages)+"\n").encode('ascii'))
        file.write((str(n_nodes)+"\n").encode('ascii'))
        file.write((str(M)+"\n").encode('ascii'))
        file.write((str(Q)+"\n").encode('ascii'))
        file.write((str(K)+"\n").encode('ascii'))
        file.write((str(pi[0])+"\n").encode('ascii'))
        file.write((str(pi[1])+"\n\n").encode('ascii'))
        file.write((str(100)+"\n\n").encode('ascii'))

       # file1.write((str(N)+"\n").encode('ascii'))
       # file1.write((str(stages)+"\n").encode('ascii'))
       # file1.write((str(n_nodes)+"\n").encode('ascii'))
       # file1.write((str(M)+"\n").encode('ascii'))
       # file1.write((str(Q)+"\n").encode('ascii'))
       # file1.write((str(K)+"\n").encode('ascii'))
       # file1.write((str(pi[0])+"\n").encode('ascii'))
       # file1.write((str(pi[1])+"\n\n").encode('ascii'))


        for k in range(1):
            A, B = tree(stages, n_nodes, matrix)

            for i in range(len(A)):
                for j in range(len(A[str(i)])):
                    file1.write((str(A[str(i)][j])+"\t").encode('ascii'))
                file1.write(('\n').encode('ascii'))   

           
            for i in range(len(B)):
                for j in range(len(A[str(i)])):
                    file1.write((str(B[str(i)][j])+"\t").encode('ascii'))
                file1.write(('\n').encode('ascii')) 

            file1.write(('\n').encode('ascii')) 

            new_matrix1 = np.zeros((n_nodes**stages, stages+1))
            new_matrix2 = np.zeros((n_nodes**stages, stages+1))

            for i in range(stages+1):
                cont = 0
                for j in range(0, n_nodes**stages):

                    if j % int(n_nodes**stages/len(A[str(i)])) == 0 and j!= 0:
                        cont = cont+1

                    new_matrix1[j][i] = A[str(i)][cont]
                    new_matrix2[j][i] = B[str(i)][cont]

            for i in range(new_matrix1.shape[0]):
                for j in range(1, new_matrix1.shape[1]):
                    file.write((str(new_matrix1[i][j])+ '\t').encode('ascii'))

                file.write(('\n').encode('ascii'))

            file.write(('\n\n').encode('ascii'))
            for i in range(new_matrix2.shape[0]):
                for j in range(1, new_matrix2.shape[1]):
                    file.write((str(new_matrix2[i][j])+ '\t').encode('ascii'))

                file.write(('\n').encode('ascii'))
            file.write(("\n\n").encode('ascii'))

        lt, Lt = liabilities(stages)
        ft, Ft = contributions(stages)

        for i in range(len(lt)):
            file.write((str(lt[i]) + '\t').encode('ascii'))
        file.write(("\n\n").encode('ascii'))
        


        for i in range(len(Lt)):
            file.write((str(Lt[i]) + '\t').encode('ascii'))
        file.write(("\n\n").encode('ascii'))
       

        for i in range(len(ft)):
            file.write((str(ft[i]) + '\t').encode('ascii'))
        file.write(("\n\n").encode('ascii'))
        

        for i in range(len(Ft)):
            file.write((str(Ft[i]) + '\t').encode('ascii'))    
        file.write(("\n\n").encode('ascii'))
        

        file.close()
    
```
